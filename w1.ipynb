```python
import numpy as np
import matplotlib.pyplot as plt

def normal_eq(xy):
    xn = [xy[i][0] for i in range(len(xy))]
    yn = [xy[i][1] for i in range(len(xy))]

    cmx = np.matrix([(1,x) for x in xn])
    my = np.matrix(yn).transpose()

    xT = cmx.transpose()

    w = ((xT*cmx)**-1)*(xT*my)

    c = w.item(0)
    m = w.item(1)

    print(f"slope:{m}        intersect:{c}")

    plt.scatter(xn, yn, color='blue')
    plt.plot(xn,[m*xi+c for xi in xn])
    plt.show()

def normal_eq_standard(xy):
    xn = [xy[i][0] for i in range(len(xy))]
    yn = [xy[i][1] for i in range(len(xy))]
    xbar = sum(xn)/len(xn)
    ybar = sum(yn)/len(yn)
    SDx = (sum([(xi-xbar)**2 for xi in xn])/(len(xn)-1))**(1/2)
    SDy = (sum([(yi-ybar)**2 for yi in yn])/(len(yn)-1))**(1/2)
    xn = [(x-xbar)/SDx for x in xn]
    yn = [(y-ybar)/SDy for y in yn]

    cmx = np.matrix([(1,x) for x in xn])
    my = np.matrix(yn).transpose()

    xT = cmx.transpose()

    w = ((xT*cmx)**-1)*(xT*my)

    c = w.item(0)
    m = w.item(1)

    print(f"slope:{m}        intersect:{c}")

    plt.scatter(xn, yn, color='blue')
    plt.plot(xn,[m*xi+c for xi in xn])
    plt.show()
    
```
